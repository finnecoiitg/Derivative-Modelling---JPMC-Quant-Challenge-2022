````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
S0,K,r,v,T1,T2  = [float(x) for x in input("Enter the values of Stock price, Strike Price, risk free rate, volatility, Timeof expiration,pretest time, no of simuls.").split()]
nsim = int(input("Enter the no of simulations:"))
flag = input("PLease enter call or option:")
````

````python
#Precompute the constant
T= (float)(T1-T2)/365     #change in time

mu = (r-0.5*v**2)*T
vr = v*np.sqrt(T)
````

#**Monte CARLO SIMULATION**

###**WHAT IS MONTE CARLO SIMULATION**



```
Monte Carlo simulations is a way of solving probabilistic problems by numerically simulating many possible scenarios so that you may calculate 
statistical properties of the outcomes, such as expectations, variances of probabilities of certain outcomes. 
In the case of Financial Derivatives, this gives us a handy tool for which to price complex derivatives for which and analytical formulae 
is not possible. 
First used by Boyle in 1977, Monte Carlo simulation provides an easy way to deal with multiple random factors and the incorporation of more 
realistic asset price processes such as jumps in asset prices.
```





````python
def op(S0,K,r,v,T,nSim,flag):
    z = np.random.standard_normal(nsim)                 #generate random numbers with mean equal 0 and std dev equal 1
    S_fwd = S0*np.exp(mu + vr*z)
    
    if flag == "call":
        price = np.maximum(S_fwd-K,0)
    elif flag == "put":
        price = np.maximum(0,K-S_fwd)
    else:
        print("error")
        return
    return np.exp(-r*T)*np.sum(price)/nsim
````

````python
call_price = op(S0,K,r,v,T,nsim,flag)
print(call_price)
````
